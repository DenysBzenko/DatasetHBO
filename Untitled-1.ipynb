```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import altair as alt
```

```python
# Loading the dataset
hbo_data = pd.read_csv('C:\\Users\\User\\Downloads\\HBO_Content.csv')

# View the first few lines of the dataset
print(hbo_data.head())
```

```python

# Delete rows with missing values in important columns
hbo_data_cleaned = hbo_data.dropna(subset=['imdb_score', 'type'])


# View the first few rows of the cleared dataset
print(hbo_data_cleaned.head())
```

```python
genre_columns = [col for col in hbo_data_cleaned.columns if 'genres_' in col]
genre_counts = hbo_data_cleaned[genre_columns].sum().sort_values(ascending=False)

# Create a bar graph
plt.figure(figsize=(12, 6))
genre_counts.plot(kind='bar')
plt.title('Number of HBO shows by genre')
plt.xlabel('Genre.')
plt.ylabel('Number of shows')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculating the average IMDb rating for each year
average_rating_per_year = hbo_data_cleaned.groupby('year')['imdb_score'].mean()

# Create a line graph
plt.figure(figsize=(12, 6))
average_rating_per_year.plot(kind='line')
plt.title('Average IMDb rating of HBO shows by year')
plt.xlabel('Year')
plt.ylabel('Average IMDb rating')
plt.grid(True)
plt.show()
```

```python
# Calculating the distribution of shows by type
type_counts = hbo_data_cleaned['rating'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of HBO shows by type')
plt.show()
```

```python
# Create a dot plot for IMDb and Rotten Tomatoes ratings
plt.figure(figsize=(10, 6))
plt.scatter(hbo_data_cleaned['imdb_score'], hbo_data_cleaned['rotten_score'])
plt.title('The relationship between IMDb and Rotten  ratings')
plt.xlabel('IMDb Rating')
plt.ylabel('Rotten Tomatoes Rating')
plt.grid(True)
plt.show()
```

```python
# Create a histogram of IMDb ratings distribution
plt.figure(figsize=(10, 6))
plt.hist(hbo_data_cleaned['imdb_score'], bins=20, edgecolor='black')
plt.title('IMDb ratings distribution for HBO shows')
plt.xlabel('IMDb Rating')
plt.ylabel('Number of shows')
plt.grid(True)
plt.show()
```
